```python
import pandas as pd
import pyspan as ps
```

```python
pd.set_option("display.max_columns",None)
file_path="/content/fifa21_raw_data.csv"
df=pd.read_csv(file_path)
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
pd.set_option("display.max_rows",None)
df.isnull().sum()
```

```python
#imputation technqiues
#custom imputation --Strings--"No Loan"
#numerical columns --mean,median,mod--
```

```python
#replace null values in loan Date End

df=ps.handle_nulls(df,"Loan Date End",action="replace",with_val="No Loan")
```

```python
df["Loan Date End"].value_counts()
```

```python
#Outlier Detection

#Outlier in Age column
#38-40
#15-17

df["Age"].value_counts()
```

```python
#IQR inter Quartile Range
#25th Percentile Q1
#50th Percentile  --median
#75th Percentile Q3

df=ps.detect_outliers(df,columns=["Age"],method="iqr",handle_missing=True)
```

```python
#team & Contract column

#\n
#start date and end date in the same column --action:split
#~

```

```python
#removing column -- photoUrl

df=ps.remove(df,operation="columns",columns="photoUrl")
df.head()
```

```python
#removing column --PlayerUrl
df=ps.remove(df,operation="columns",columns="playerUrl")
df.head()
```

```python
#cleaning weight column

df = ps.remove_chars(df, ['Weight'], custom_characters='lbs')
df.head()
```

```python
#cleaning columns that contain stars
df = ps.remove_chars(df, ['W/F','SM','IR'], custom_characters='★')
df.head()
```

```python
#cleaning Hits column
df = ps.remove_chars(df, ['Hits'], custom_characters='\n')
df.head()
```

```python
#spliting positions column
df=ps.split_column(df,column="Positions",delimiter=" ")
df.head()
```

```python
#splitting "Team&Contract" --team,contract start date and end date

df=ps.remove_chars(df,["Team & Contract"],custom_characters="\n")
df.head()
```

```python
df=ps.split_column(df,column="Team & Contract",delimiter=" ~ ")
df.head()
```

```python
df["Team Name"]=df["Team & Contract_1"].str[:-4]
df.head()
```

```python
df["Start Contract"] =df["Team & Contract_1"].str[-4:]
df.head()
```

```python
df["End Contract"] = df["Team & Contract_2"]
df.head()
```

```python
#Deleting Column -- Team & Contract_1, Team & Contract_2
df=ps.remove(df,operation="columns",columns=["Team & Contract_2"])
df.head()
```
